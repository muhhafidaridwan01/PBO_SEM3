# Analisis Produk

**Distribusi berdasarkan jumlah pesanan(order)**

```python
ax = df['Quantity'].plot.box(
    showfliers=False,
    grid=True,
    figsize=(8, 5)
)

ax.set_ylabel('Order Quantity')
ax.set_title('Quantity Distribution')

plt.suptitle("")
plt.show()
```

**Rincian Data Berdasarkan Jumlah**

```python
pd.DataFrame(df['Quantity'].describe())
```

**Jumlah Pesanan Berdasarkan Tanggal**

```python
df.loc[df['Quantity'] > 0].shape
```

```python
df = df.loc[df['Quantity'] > 0]
```

```python
monthly_orders_df = df.set_index('InvoiceDate')['InvoiceNo'].resample('ME').nunique()
```

```python
monthly_orders_df
```

```python
ax = pd.DataFrame(monthly_orders_df.values).plot(
    grid=True,
    figsize=(10,7),
    legend=False
)

ax.set_xlabel('date')
ax.set_ylabel('number of orders/invoices')
ax.set_title('Total Number of Orders Over Time')

plt.xticks(
    range(len(monthly_orders_df.index)), 
    [x.strftime('%m.%Y') for x in monthly_orders_df.index], 
    rotation=45
)

plt.show()
```

**Total Penjualan Berdasarkan Tanggal**

```python
df['Sales'] = df['Quantity'] * df['UnitPrice']
```

```python
monthly_revenue_df = df.set_index('InvoiceDate')['Sales'].resample('ME').sum()
```

```python
monthly_revenue_df
```

**Grafik Total Penjualan berdasarkan tanggal**

```python
ax = pd.DataFrame(monthly_revenue_df.values).plot(
    grid=True,
    figsize=(10,7),
    legend=False
)

ax.set_xlabel('date')
ax.set_ylabel('sales')
ax.set_title('Total Revenue Over Time')

ax.set_ylim([0, max(monthly_revenue_df.values)+100000])

plt.xticks(
    range(len(monthly_revenue_df.index)), 
    [x.strftime('%m.%Y') for x in monthly_revenue_df.index], 
    rotation=45
)

plt.show()
```
